```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', encoding='unicode-escape')
```

```python
df.info()
```

```python
df.head()
```

```python
for col in df.columns:
  df[col] = df[col].replace('\xa0', pd.NA)  # Use pd.NA for null values

# Drop rows with null values (after replacement)
df_filtered = df.dropna(how='any')

# Print the modified DataFrame
print(df_filtered)
```

```python
df_filtered.info()
```

```python
df_filtered['ct1_agent'] = df_filtered['ct1_agent'].astype('string')
```

```python
df_filtered.info()
```

```python
df = df_filtered
```

```python
# group similar datatype columns (for data type conversion)

attr_list = ['ct1_R','ct1_ACS','ct1_ADR',
             'ct2_R','ct2_ACS','ct2_ADR',
             'ct3_R','ct3_ACS','ct3_ADR',
             'ct4_R','ct4_ACS','ct4_ADR',
             'ct5_R','ct5_ACS','ct5_ADR',
             't1_R','t1_ACS','t1_ADR',
             't2_R','t2_ACS','t2_ADR',
             't3_R','t3_ACS','t3_ADR',
             't4_R','t4_ACS','t4_ADR',
             't5_R','t5_ACS','t5_ADR']

percent_list = ['ct1_KAST', 'ct1_HS',
                'ct2_KAST', 'ct2_HS',
                'ct3_KAST', 'ct3_HS',
                'ct4_KAST', 'ct4_HS',
                'ct5_KAST', 'ct5_HS',
                't1_KAST', 't1_HS',
                't2_KAST', 't2_HS',
                't3_KAST', 't3_HS',
                't4_KAST', 't4_HS',
                't5_KAST', 't5_HS']

agent_list = ['ct1_agent', 'ct2_agent', 'ct3_agent', 'ct4_agent', 'ct5_agent',
              't1_agent', 't2_agent', 't3_agent', 't4_agent', 't5_agent']

other_list = ['map', 'winner_team']
```

```python
# datatype conversion

for col in attr_list:
    df[col] = df[col].astype(float)
    
for col in percent_list:
    df[col] = df[col].astype('string')
    
    # convert percentage data to float
    df[col] = (pd.to_numeric(df[col].str.strip('%'), errors='coerce') / 100)
    
for col in agent_list:
    df[col] = df[col].astype('string')
    
for col in other_list:
    df[col] = df[col].astype('string')
```

```python
df.info()
```

```python
df.head()
```

```python
# remove the "PICK" from map

# 1) check for the unique data
df['map'].unique()

# 2) remove unwanted string
strings_to_remove = ["PICK", "\r", "\n"]
df['map'] = df['map'].str.replace('|'.join(strings_to_remove), '', regex=True)
```

```python
# store preprocessed data into csv.

df.to_csv('data_cleaned.csv', index=False)
```
